```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
```

```python
lungcancer = pd.read_csv("lung_cancer.csv")
```

```python
lungcancer
```

```python
issmokingyes = lungcancer[lungcancer["SMOKING"]==2]
```

```python
issmokingyes
```

```python
plt.figure()
plt.title("SMOKING")
plt.xlabel("SMOKING FACTOR")
plt.ylabel("CANCER ?")

plt.plot(lungcancer.SMOKING,lungcancer.LUNG_CANCER) 
plt.savefig("smokingandcancer.jpg",facecolor="green")
plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(lungcancer.AGE,lungcancer.LUNG_CANCER,color="red") 

plt.title("AGE AND CANCER TOTAL")
plt.savefig("ageandcancertotal.jpg",facecolor="blue")
plt.show()
```

```python
y = lungcancer[lungcancer["GENDER"]=="F"]
```

```python
plt.figure(figsize=(12,6))

plt.plot(x.AGE,x.LUNG_CANCER,color="green") 

plt.plot(y.AGE,y.LUNG_CANCER,color="pink")

plt.xlabel("AGES")

plt.ylabel("BOY AND WOMAN SURVEY")

plt.title("TOTAL SURVEY PEOPLE CANCER")

plt.savefig("boy-womansurvey.jpg",facecolor="green")
plt.show()
```

```python
xc = x[x["LUNG_CANCER"]=="YES"]
```

```python
yc = y[y["LUNG_CANCER"]=="YES"]
```

```python
yc
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)   
plt.plot(xc.AGE,xc.PEER_PRESSURE,color="r") 
plt.xlabel("AGE")
plt.ylabel("PEER PRESSURE")
plt.title("PEER PRESSURE EFFECTS ON MAN")

plt.subplot(2,2,2)
plt.plot(yc.AGE,yc.PEER_PRESSURE,color="blue")
plt.xlabel("AGE")
plt.ylabel("PEER PRESSURE")
plt.title("PEER PRESSURE EFFECTS ON WOMAN")
plt.savefig("peerpressure.jpg",facecolor="pink")
plt.show()
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)   
plt.plot(xc.AGE,xc.COUGHING,color="r") 
plt.xlabel("AGE")
plt.ylabel("COUGHING")
plt.title("COUGHING EFFECTS ON MAN")

plt.subplot(2,2,2)
plt.plot(yc.AGE,yc.COUGHING,color="blue")
plt.xlabel("AGE")
plt.ylabel("COUGHING")
plt.title("COUGHING EFFECTS ON WOMAN")
plt.savefig("couhgıngeffects.jpg",facecolor="green")
plt.show()
```
